```python
import requests
from bs4 import BeautifulSoup
# Function to scrape BBC Good Food

ingredients = ['flageolet','sausage']

def scrape_recipes(ingredients):
    # search_url = f"https://www.bbcgoodfood.com/search/recipes?q={'%20'.join(ingredients)}"
    search_url = f"https://www.bbcgoodfood.com/search?q={'%20'.join(ingredients)}"
    response = requests.get(search_url)
    print(response)
    if response.status_code != 200:
        return "Error fetching recipes from BBC Good Food. Please try again later."
    
    soup = BeautifulSoup(response.content, "html.parser")
    recipes = []
    print(recipes)
    
    # Extract recipe titles and links
    for item in soup.find_all('article', class_='standard-card-new'):
        title_tag = item.find('h4', class_='standard-card-new__display-title')
        link_tag = item.find('a', class_='standard-card-new__article-title')
        if title_tag and link_tag:
            title = title_tag.text.strip()
            link = "https://www.bbcgoodfood.com" + link_tag['href']
            recipes.append(f"{title}: {link}")
    
    return recipes[:5]  # Return top 5 recipes

scrape_recipes(ingredients)
```
